```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("Data/AS_Data.xlsx", engine='openpyxl')
```

```python
x_ax = np.linspace(1,90,90)
plt.figure(figsize=(15,5))
plt.plot(x_ax, df.error, color='black', marker='.' )
plt.axhline(0, color='grey', linestyle='--')
plt.axvline(9.5, color='grey', linestyle='--')
plt.axvline(29.5, color='grey', linestyle='--')
plt.axvline(38.5, color='grey', linestyle='--')
plt.axvline(69.5, color='grey', linestyle='--')
plt.title('End point error', fontsize=16)
plt.xlabel('Trial #', fontsize=16)
plt.ylabel('Error (degrees)', fontsize=16)
plt.xlim([1, 90])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(x_ax, df.error, color='black', marker='.' )
ax.set_xlabel("Trial #", fontsize=16)
ax.set_ylabel("Error (degrees)", fontsize=16)
ax.set_title('End Point Error', fontsize=16)
ax.axhline(0, color='black', linestyle='--', alpha=0.3)
ax.axvline(9.5, color='black', linestyle='--', alpha=0.3)
ax.axvline(29.5, color='black', linestyle='--', alpha=0.3)
ax.axvline(38.5, color='red', linestyle='--', alpha=0.6)
ax.axvline(69.5, color='black', linestyle='--', alpha=0.3)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.text(3, 15, 'Practice')
ax.text(18, 15, 'Baseline')
ax.text(50, 15, 'Exposure')
ax.text(78, 15, 'Post-Exposure')
ax.text(38.5, 0, 'Rotation Introduced', rotation=270)
ax.set_xlim([0.8, 90.2])

```
